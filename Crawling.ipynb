```python
import requests
from bs4 import BeautifulSoup
import re

# 기본 URL
base_url = "https://nces.ed.gov/collegenavigator/?s=all&ic=1&pg="

# 결과 저장
school_ids = []

# 페이지 반복
for page_num in range(1, 35):  # 1~34 페이지
    # 페이지 URL 생성
    url = f"{base_url}{page_num}"
    print(f"Requesting page: {url}")
    
    # 페이지 요청
    response = requests.get(url)
    
    if response.status_code == 200:
        # HTML 파싱
        soup = BeautifulSoup(response.text, 'html.parser')
        
        # 학교 링크 추출
        links = soup.find_all('a', href=True)  # 모든 링크 태그 가져오기
        
        for link in links:
            href = link['href']
            # ID가 포함된 링크 필터링
            match = re.search(r'id=(\d+)', href)
            if match:
                school_id = match.group(1)
                if school_id not in school_ids:  # 중복 제거
                    school_ids.append(school_id)
                    print(f"Found school ID: {school_id}")
    else:
        print(f"Failed to retrieve page {page_num}. Status code: {response.status_code}")

# 결과 출력 및 저장
print(f"Total School IDs Found: {len(school_ids)}")
print(school_ids)

# 파일에 저장
with open('school_ids.txt', 'w') as file:
    for school_id in school_ids:
        file.write(f"{school_id}\n")
```
